```python
df = pd.read_csv('mergedClimateChange.csv', index_col = 0)
```

```python
df
```

```python
df_copy = df.copy()
df_copy["Temp"] = df_copy["Temperature"]
df_copy.drop(["Temperature", "Year"], axis = 1, inplace=True)
df_copy
```

```python
X2 = df_copy.iloc[:,: -1]
y2 = df_copy.iloc[:, -1]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state=0)
reg = LinearRegression(fit_intercept=False)
reg.fit(X2_train, y2_train)
y2_pred = reg.predict(X2_test)
```

```python
print(f"mse: {metrics.mean_squared_error(y2_test, y2_pred)}")
print(f"r^2: {metrics.r2_score(y2_test, y2_pred)}")
```

## We used linear regression, as we want to use X different numeric data to obtain the y outcome
## X = allegedly causes of the global warming
## y = the temperature we want to predict

```python
X = df[df.columns[(df.columns != 'Year') & (df.columns != 'Temperature')]]
y = df['Temperature']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
reg = LinearRegression(fit_intercept=False)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

```python
print(f"mse: {metrics.mean_squared_error(y_test, y_pred)}")
print(f"r^2: {metrics.r2_score(y_test, y_pred)}")
```
